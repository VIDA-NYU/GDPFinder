```python
import cenpy
from cenpy import products
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generate visualization of houshold income (or other metric) by census tract for a city
ny = products.ACS(2019).from_place('New York, NY', level='tract',
                                        variables=['B00002*', 'B19013_001E'])

f, ax = plt.subplots(1,1,figsize=(10,10))
ny.dropna(subset=['B19013_001E'], axis=0).plot('B19013_001E', ax=ax, cmap='plasma')
ax.set_facecolor('k')
```

```python
# show outline of city with census tracts

dectest = products.Decennial2010()
tuc, tuc_bounds = dectest.from_place('New York, NY', 
                                     level='tract',
                                     return_bounds=True, 
                                     strict_within=False)

f, ax = plt.subplots(1,1,figsize=(10,10))
tuc.plot(edgecolor='slategrey', color='none', ax=ax)
tuc_bounds.plot(edgecolor='cornsilk', color='none', ax=ax, linewidth=2)
ax.set_facecolor('k')
```
